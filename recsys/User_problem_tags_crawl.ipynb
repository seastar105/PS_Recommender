```python
import os
import json
import requests
import time
```

```python
base_url = "https://solved.ac/api/v3"
api_string = "user/problem_tag_stats?handle="
user_directory = "solved_data/user"
stats_directory = "solved_data/user_tags"
sleep_seconds = 200
```

```python
def call_file_sorted(path):
    directory = os.fsencode(path)
    ret = []
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        filename = os.path.join(path, filename)
        ret.append(filename)
    ret.sort()
    return ret

def change_path(target_path, source_path, ext):
    basename = os.path.basename(source_path)
    basename = os.path.splitext(basename)[0]
    return_path = os.path.join(target_path, basename) + ext
    return return_path
```

```python
def get_user_tag_stats(user_handle):
    url = base_url + "/" + api_string + user_handle
    r = requests.get(url)
    if r.status_code != requests.codes.ok:
        # if too many requests return 
        if r.status_code == 429:
            time.sleep(sleep_seconds) # sleep for some time
            return get_user_tag_stats(user_handle)
        else :
            print(r.status_code)
            print(r.text)
            time.sleep(sleep_seconds)
            return get_user_tag_stats(user_handle)
    return r.json()
```

```python
user_json_list = call_file_sorted(user_directory)

for user_json in user_json_list[23467:]:
    time.sleep(2)
    with open(user_json, "r", encoding="utf-8") as f:
        handle = json.load(f)['handle']
    res = get_user_tag_stats(handle)
    stat_path = change_path(stats_directory, user_json, ".json")
    with open(stat_path, "w", encoding="utf-8") as f:
        json.dump(res, f, ensure_ascii=False)
    cnt += 1
    if cnt % 100 == 0:
        print(cnt, "Done")
```

```python
user_json_list = call_file_sorted(user_directory)
cnt = 0
for user_json in user_json_list:
    with open(user_json, "r", encoding="utf-8") as f:
        handle = json.load(f)['handle']
    stat_path = change_path(stats_directory, user_json, ".json")
    with open(stat_path, "r", encoding="utf-8") as f:
        stats = json.load(f)
    stats['handle'] = handle
    cnt += 1
    with open(stat_path, "w", encoding="utf-8") as f:
        json.dump(stats, f, ensure_ascii=False)
```
