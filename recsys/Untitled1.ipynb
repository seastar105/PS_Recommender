```python
def call_file_sorted(path):
    directory = os.fsencode(path)
    ret = []
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        filename = os.path.join(path, filename)
        ret.append(filename)
    ret.sort()
    return ret

def change_path(target_path, source_path, ext):
    basename = os.path.basename(source_path)
    basename = os.path.splitext(basename)[0]
    return_path = os.path.join(target_path, basename) + ext
    return return_path
```

```python
cnt = 0
tot = 0
user_solved_dir = "solved_data/user_solved"
user_list = call_file_sorted(user_solved_dir)
for user_solved in user_list:
    with open(user_solved, "r", encoding="utf-8") as f:
        tot += len(json.load(f)["problems"])
    cnt += 1
    if cnt % 100 == 0:
        print(cnt, tot)
print(cnt, tot)
```
